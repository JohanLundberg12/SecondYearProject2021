# Word sentiment data frame

_The word counts used are only from the training file_

_also note that the word counts are based on sentences that have been lower cased. Also, I have removed punctuation and stopwords from the counts_

## Importing libraries and word count dictionaries

I made the dictionaries in another file. In the process of making these, I had to run multiple cells that took around an hour to finish. Therefore, I saved the dictionaries in seperate files, such that I did not have to run the other notebook again

```python
import json
import pandas as pd
```

```python
word_count = json.loads(open("word_counts/word_count.json", "r").read())
positive_count = json.loads(open("word_counts/positive_word_count.json", "r").read())
negative_count = json.loads(open("word_counts/negative_word_count.json", "r").read())
```

## Generating word count data frame

```python
vocab = set()
for word in word_count:
    vocab.add(word)
```

```python
bestlist = list()
for word in vocab:
    if (word in negative_count) and (not word in positive_count):
        bestlist.append([word, 'negative', word_count[word], negative_count[word], 0, 0])
    elif (not word in negative_count) and (word in positive_count):
        bestlist.append([word, 'positive', word_count[word], 0, positive_count[word], 1])
    else:
        if negative_count[word] < positive_count[word]:
            bestlist.append([word, 'positive', word_count[word], negative_count[word], positive_count[word], positive_count[word]/word_count[word]])
        else:
            bestlist.append([word, 'negative', word_count[word], negative_count[word], positive_count[word], positive_count[word]/word_count[word]])
```

```python
word_label_df = pd.DataFrame(bestlist, columns = ['word', 'sentiment', 'word count', 'negative count', 'positive count', 'positive frequency'])
word_label_df
```

## Finding hard words

I define a hard word to be a word where the positive frequency (how often the word is in a sentence with positive sentiment, when it appears) is between 0.4 and 0.6.

I also filtered out words that appear less than 50 times

```python
hard_df = word_label_df[(word_label_df['positive frequency'] <= 0.6) & (word_label_df['positive frequency'] >= 0.4) & (word_label_df['word count'] >= 50)]
hard_df
```
