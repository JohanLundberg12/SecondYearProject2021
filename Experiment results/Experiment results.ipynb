```python
import pandas as pd
```

```python
beauty = pd.read_csv("hardcases_beauty_cheatsheet.csv")
```

```python
from sklearn.metrics import f1_score, accuracy_score
```

```python
music = pd.read_csv("hardcases_music_cheatsheet.csv")
```

```python
sports = pd.read_csv("hardcases_sports_cheatsheet.csv")
```

```python
def lower(string):
    return string.lower()
```

```python
beauty["Phillip"] = pd.read_excel("hardcases_beauty_phillip.xlsx")["sentiment"]
beauty["Signe"] = pd.read_excel("hardcases_beauty_Signe.xlsx")["sentiment"].apply(lower)
beauty["Daniel"] = pd.read_excel("hardcases_beauty-done-Daniel.xlsx")["sentiment"]
```

```python
beauty = beauty.replace({"positive":1.0,"negative":0.0})
```

```python
music["Hanne"] = pd.read_excel("hardcases_music_hanne.xlsx")["sentiment"].replace({"positiv":"positive","negativ":"negative"})
music["Johan"] = pd.read_excel("hardcases_music_Johan.xlsx")["sentiment"].replace({"positiv":"positive","negativ":"negative"})
```

```python
music = music.replace({"positive":1.0,"negative":0.0})
```

```python
sports["Andreas"] = pd.read_excel("hardcases_sports_Andreas.xlsx")["sentiment"].replace({"positiv":"positive","negativ":"negative"})
sports["Mae"] = pd.read_excel("hardcases_sports-1_mae.xlsx")["sentiment"]
sports["Amalie"] = pd.read_excel("hardcase_sport_amalie.xlsx")["sentiment"].replace({"positiv":"positive","negativ":"negative"})
sports["Casper"] = pd.read_excel("hardcases_sports-1-Casper.xlsx")["sentiment"].replace({"positiv":"positive","negativ":"negative"})
sports["Bjarne"] = pd.read_excel("hardcases_sports_B_1.xlsx")["sentiment"].replace({"positiv":"positive","negativ":"negative"})
```

```python
sports = sports.replace({"positive":1.0,"negative":0.0})
```

## Amalie, Casper, Bjarne og Daniel

```python
def get_acc(df):
    for part in list(df.columns[4:]):
        print("Score for ", part)
        print("Overall accuracy: ", accuracy_score(df[part],df["sentiment"]) )
        
        for case in set(sports["casetype"]):
            temp_df = df[df["casetype"] == case]
            print(case, " accuracy: ", accuracy_score(temp_df[part],temp_df["sentiment"]))
```

```python
get_acc(sports)
```

```python
get_acc(beauty)
```
